### Review Count Thresholds

````python
print("REVIEW‑COUNT THRESHOLDS")
df["rev_bin"] = pd.cut(df.number_of_reviews, [0, 10, 30, 50, 100, np.inf],
                       labels=["0‑10", "11‑30", "31‑50", "51‑100", "100+"])
rev_stats = (df.groupby("rev_bin")
               .agg(listings=("id", "size"),
                    median_price=("price", "median"),
                    median_avail=("availability_365", "median")))
print(rev_stats, "\n")
````

### Price Band VS Review

````python
print("PRICE BANDS VS. REVIEW VELOCITY")
price_bins = pd.cut(df.price, [0, 60, 80, 150, 250, 400, np.inf],
                    labels=["≤60", "61‑80", "81‑150", "151‑250",
                            "251‑400", ">400"])
band = (df.groupby(price_bins)
          .agg(listings=("id", "size"),
               median_reviews_pm=("reviews_per_month", "median"),
               median_avail=("availability_365", "median")))
print(band, "\n")
````

### Seasonal Review Distribution

````python
print("SEASONAL REVIEW DISTRIBUTION (2018‑19 reviews only)")
df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")
recent_reviews = df.dropna(subset=["last_review"])
reviews_1819 = recent_reviews[recent_reviews.last_review.dt.year >= 2018]
month_counts = reviews_1819.last_review.dt.month.value_counts().sort_index()
month_pct = 100 * month_counts / month_counts.sum()
print(month_pct.rename("share_%"), "\n")
````

### Peak Season Share

#### Chart - 7 **Price Distribution by Neighbourhood**

````python
plt.figure(figsize=(10, 6))
boroughs = sorted(df["neighbourhood_group"].unique())
palette  = sns.color_palette("rocket_r", len(boroughs))

fig, ax = plt.subplots(len(boroughs), 1, sharex=True,
                       figsize=(9, 7), dpi=100)

for i, b in enumerate(boroughs):
    sns.kdeplot(
        df.loc[df["neighbourhood_group"] == b, "price"].clip(upper=500),
        ax=ax[i], fill=True, bw_adjust=.8, color=palette[i], alpha=.9,
        clip=(0, 500)
    )
    ax[i].set_ylabel(b, rotation=0, labelpad=30, fontsize=11)
    ax[i].set_yticks([])

ax[-1].set_xlabel("Price ($, capped at 500)")
fig.suptitle("Ridgeline Plot – Price Distribution by Borough", y=.93)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````
